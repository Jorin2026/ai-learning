```python
import os  
#用os库创建csv文件，存储焊接参数
os.makedirs(os.path.join('..', 'data'), exist_ok=True)
data_file = os.path.join('..', 'data', 'welding.csv')

with open(data_file, 'w') as f:
    f.write('电流,电压,压力,焊接质量\n')
    f.write('150,220,3.5,良好\n')
    f.write('160,NA,3.8,良好\n')
    f.write('145,215,NA,不良\n')
    f.write('170,230,4.0,良好\n')
```

```python
import pandas as pd 
#用pandas 读取csv发现缺失值NaN
data = pd.read_csv(data_file)
print(data)
```

```python
#用每列的平方值填充缺失值
inputs = data.iloc[:,0:3]  #取前3列（电流，电压，压力）
inputs = inputs.fillna(inputs.mean())
print(inputs)
```

```python
import torch  用平均值NaN再转成张量给模型用
x=torch.tensor(inputs.values)
print(x)
```

```python
#模拟3条焊接记录，每行是【电流，电压，压力】
data = torch.tensor([[150.0,220.0,3.5],
                    [160.0,225.0,3.8],
                    [145.0,215.0,3.6]])

#权重：电流权重0.6 电压权重0.3 压力权重0.1 
weight = torch.tensor([0.6,0.3,0.1])

print(data.shape)
print(weight.shape)
```

```python
result = data * weight 
print(result)
```

```python
score = (data * weight).sum(dim = 1)
print(score)
```
